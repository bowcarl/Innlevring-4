## Gjennomsnitt og Standardavvik

```python
from pylab import *

solflekker = loadtxt("datatext.csv", skiprows=1, delimiter=",")

nr = solflekker[:,0]

SUNACTIVITY = solflekker[:,1]

plot(nr, SUNACTIVITY, color = "red")
grid()
title("Solflekker per år")
xlabel(' år ')
ylabel(' antall solflekker ')
show()

snitt = mean(SUNACTIVITY)
avvik = std(SUNACTIVITY)

print("gjennomsnitt:", snitt,"solflekker | standardavvik:", avvik,"solflekker")

```

###### Gjennomsnitt : 
Gjennomsittet av $n$ målinger er definert som summen av alle målingene dividert på $n$:

$$\bar{x} =\frac{1}{n}\sum_{i=1}^n x_i$$

###### Standardavvik : 

Standardavviktet til $n$ målinger er definert slik:

$$s = \sigma^2=\sqrt{\frac{1}{n}\sum(x_i-\bar{x})^2}$$

## Regresjon

```python
from pylab import *
from sklearn.metrics import r2_score

T = array([0, 20, 40, 60, 80, 100])
sol_NH3 = array([88.5, 56.0, 34.0, 20.0, 11.0, 7.0])
sol_NaCl = array([35.7, 35.9, 36.4, 37.1, 38.0, 39.2])


#regrisjon 1

grad = 3
reg_NaCl = polyfit(T, sol_NaCl, grad )

x = linspace(0, 250, 100000)
y = reg_NaCl[0]*x**3+reg_NaCl[1]*x**2+reg_NaCl[2]*x+reg_NaCl[3]
y_R2 = reg_NaCl[0]*T**3+reg_NaCl[1]*T**2+reg_NaCl[2]*T+reg_NaCl[3]
R2 = r2_score(sol_NaCl, y_R2)

print("regrisjonen er" , R2*100 , "% nøyaktig")


scatter(T, sol_NaCl)

plot(x, y, color="red", label = "NaCl")
grid()
title("Natriumklorid")
xlabel("temperatur")
ylabel("løslighet")
legend()
show()

# regrisjon 2

grad = 3
reg_NH3 = polyfit(T, sol_NH3, grad)

x = linspace(0, 250, 100000)
y = reg_NH3[0]*x**3+reg_NH3[1]*x**2+reg_NH3[2]*x+reg_NH3[3]
y2 = reg_NH3[0]*T**3+reg_NH3[1]*T**2+reg_NH3[2]*T+reg_NH3[3]
R12 = r2_score(sol_NH3, y2)


scatter(T, sol_NH3)

plot(x, y, color="green", label = "$NH_3$")
grid()
title("Ammoniakk")
xlabel("temperatur")
ylabel("løslighet")
ylim(0,70) #ser bort ifra negativ løslighet
legend()
show()

print("regrisjonen er" , R12*100 , "% nøyaktig")


```
